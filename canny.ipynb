```python
import cv2
import numpy as np
```

```python
gray = cv2.imread('ex2.bmp')
img_gaussian = cv2.GaussianBlur(gray,(3,3),0)
```

```python
img_canny = cv2.Canny(gray,60,120)
point1=0
point2=0
point3=0
point4=0
x1=0
x2=0
x3=0
x4=0
y1=0
y2=0
y3=0
y4=0
```

```python
for i in range(0,180):
	for j in range(0,50):
		if img_canny[i][j]==255:
			if i<90:
				if j>24:
					x1 += i;
					y1 += j;
					point1 += 1;
				else : 
					x2 += i;
					y2 += j;
					point2 += 1;
			else :
				if j>24:
					x3 += i;
					y3 += j;
					point3 += 1;
				else : 
					x4 += i;
					y4 += j;
					point4 += 1; 
			
			
			
						
			
centroidx1 = x1/point1;
centroidy1 = y1/point1;
```
